```python
import time
import pandas as pd
from urllib.request import Request, urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup
```

```python
def scrape_berita(keyword, jumlah_berita):
    keyword = keyword.replace(' ', '+')
    judul_list, konten_list, date_list, link_list = [], [], [], []
    page = 1
    total_berita = 0
    while len(judul_list) < jumlah_berita:
        link = f'https://www.detik.com/search/searchall?query={keyword}&page={page}'
        UserAgent = Request(link, headers={'User-Agent': 'Mozilla/5.0'})
        try:
            print(f"[INFO] Mengunduh halaman {page}...")
            html_text = urlopen(UserAgent)
            time.sleep(1)
            soup = BeautifulSoup(html_text, "html.parser")
            daftar = soup.find_all("article", {"class": "list-content__item"})
            for element in daftar:
                if len(judul_list) >= jumlah_berita:
                    break
                judul_elem = element.find('h3', class_='media__title')
                print(f"\n[DEBUG] Elemen judul yang ditemukan: {judul_elem}")               

                # Ambil judul, konten, dan tanggal
                judul = judul_elem.get_text(strip=True) if judul_elem else 'Judul tidak ditemukan'
                konten = element.find('div', class_='media__desc').get_text(strip=True) if element.find('div', class_='media__desc') else 'Konten tidak ditemukan'
                
                # Coba beberapa tag atau kelas yang berbeda untuk elemen tanggal
                date_elem = (
                    element.find('time') or
                    element.find('span', class_='date') or
                    element.find('div', class_='tanggal') or
                    element.find('p', class_='publish-time')  # tambahkan sesuai struktur HTML sebenarnya
                    )
                date = date_elem.get_text(strip=True) if date_elem else 'Tanggal tidak ditemukan'

                
                # Link berita
                link_berita = judul_elem.find('a')['href'] if judul_elem and judul_elem.find('a') and 'href' in judul_elem.find('a').attrs else 'Link tidak ditemukan'
                
                # Append data ke list
                judul_list.append(judul)
                konten_list.append(konten)
                date_list.append(date)
                link_list.append(link_berita)
                
                total_berita += 1
                print(f"[PROGRESS] Berita ke-{total_berita}: {judul} - Tanggal: {date}")

            page += 1
            
        except HTTPError as e:
            print(f"HTTPError: {e.code} - {e.reason}")
            break

    # Konversi ke DataFrame
    data = pd.DataFrame({
        'Judul': judul_list,
        'Konten': konten_list,
        'Tanggal': date_list,
        'Link': link_list,    
    })
    
    return data

keyword = "Gen z"
jumlah_berita = 100
hasil_scrape = scrape_berita(keyword, jumlah_berita)

if hasil_scrape is not None:
    print("\n[INFO] Proses scraping selesai. Berikut hasilnya:\n")
    print(hasil_scrape)
```

```python
df = pd.DataFrame(hasil_scrape)
```
