```python
hasil = pd.concat([data_hasil['Judul'], data_hasil['Link']], axis=1)
hasil.columns = ['Judul', 'Link']
```

```python
hasil.head()
```

# lower case

```python
hasil['judul_lower_case'] = hasil['Judul'].apply(lambda x: x.lower() if isinstance(x, str) else x)
hasil.head()
```

# Removal of URLs and HTML tags

```python
import re
from bs4 import BeautifulSoup
def remove_html_tags(text):
    text = re.sub(r'http\S+|www.\S+', '', text)
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()
hasil['remove_html'] = hasil['judul_lower_case'].apply(remove_html_tags)
```

```python
hasil.head()
```

# Remove special characters

```python
import re
def remove_special_characters(text):
    text = re.sub(r'[^A-Za-z\s\n]', '', text)
    return text
hasil['remove_special_characters'] = hasil['remove_html'].apply(remove_special_characters)
```

```python
hasil.head()
```

# Remove escape characters

```python
def remove_escape_characters(text):
    return re.sub(r'[\n\r\t]', ' ', text)
hasil['remove_escape_characters'] = hasil['remove_special_characters'].apply(remove_escape_characters)
```

```python
hasil.head()
```

# Remove extra spaces

```python
def normalize_whitespace(text):
    return ' '.join(text.split())
hasil['cleaned_space'] = hasil['remove_escape_characters']
```

```python
hasil.head()
```

# Tokenisasi

```python
hasil['tokenisasi'] = hasil['cleaned_space'].apply(word_tokenize)
hasil.head()
```

# Menghapus Stopword

```python
nltk.download('stopwords')
stop_words_english = set(stopwords.words('english'))
stop_words_indonesia = set(stopwords.words('indonesian'))
def remove_stopwords_english(tokens):
    return [word for word in tokens if word.lower() not in stop_words_english]
def remove_stopwords_indonesia(tokens):
    return [word for word in tokens if word.lower() not in stop_words_indonesia]
hasil['remove_stopword_english'] = hasil['tokenisasi'].apply(remove_stopwords_english)
hasil['remove_stopword_indonesia'] = hasil['remove_stopword_english'].apply(remove_stopwords_indonesia)
```

```python
class TalaStemmer:
    def __init__(self):
        self.prefixes = [
            'me', 'mem', 'men', 'meng', 'mend', 'meny',
            'ber', 'per', 'pe', 'di', 'se', 'ke', 'ter', 'di'
        ]
        self.suffixes = [
            'kan', 'i', 'an', 'lah', 'kah', 'ku', 'mu', 'nya', 's', 'es', 'in'
        ]
    def remove_prefix(self, word):
        for prefix in self.prefixes:
            if word.startswith(prefix):
                return word[len(prefix):]
            return word
    def remove_suffix(self, word):
        for suffix in self.suffixes:
            if word.endswith(suffix):
                return word[:-len(suffix)]
            return word
    def stem(self, words):
        return [self.remove_suffix(self.remove_prefix(word)) for word in words]
tala_stemmer = TalaStemmer()
hasil['data_post_cleaned'] = hasil['remove_stopword_indonesia'].apply(tala_stemmer.stem)
hasil
```
